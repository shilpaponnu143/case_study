# 3. Plot a bar graph of male vs female buyers participated in the sales.


```python
import matplotlib.pyplot as plt
M=(data['Buyer Gender']=='Male').sum() #Finding total no of male buyers
M
```

```python
F=(data['Buyer Gender']=='Female').sum() #Finding total no of female buyers
F
```

```python
x=["Male",'Female']
Height=[M,F]
plt.bar(x,Height)
plt.xlabel("Gender")
plt.ylabel("Count")
plt.title("Male vs Female Buyers") #Plotting bar graph of male vs female buyers
```

# 4. Find the top 5 cars based on their sales price.


```python
data.nlargest(5,'Sale Price')[['Make','Model','Sale Price']] #Finding top 5 cars corresponding to their sales price
```
